# <span style="color:rgb(145, 255, 172);">Predictive Modelling</span>

## Importing LogisticRegression, train_test_split and StandardScaler 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

## Spliting dataset for training and testing

```python
y = df['attack']
x = important_features
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
```

## Scaling xtrain and xtest dataset

```python
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

## Create an LogisticRegression Model Instance

```python
lor = LogisticRegression()
```

## Fitting the regression model to the data

```python
lor.fit(pd.DataFrame(xtrain), ytrain)
```

## Making predictions using the xtest dataset

```python
y_pred = lor.predict(pd.DataFrame(xtest))
y_pred
```

# <span style="color:rgb(145, 255, 172);">Model Evaluation</span>

## Evaluation

### Importing roc_curve and area under the curve

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
```

### Calculating predicted probabilities, false positive rates (FPR), true positive rates (TPR) and thresholds

```python
y_probs = lor.predict_proba(xtest)[:, 1]
fpr, tpr, thresholds = roc_curve(ytest, y_probs)
```

### Calculating area under the curve

```python
roc_auc = auc(fpr, tpr)
roc_auc
```

### Visualizing predictions

```python
# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(ytest, y_pred)
```

### Classification Report

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, y_pred))
```

### Balancing Class

```python
df_enhanced['attack'] = df['attack']
df_enhanced
```

```python
# Dropping Duplicated 

df_enhanced = df_enhanced.drop_duplicates()
df_enhanced
```

```python
from sklearn.utils import class_weight

# Check class distribution in the target column
class_distribution = df_enhanced['attack'].value_counts()

# Calculate class weights for balancing
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=df_enhanced['attack'].unique(),
    y=df_enhanced['attack']
)

# Display class distribution and weights
class_weights_dict = dict(zip(df_enhanced['attack'].unique(), class_weights))

class_weights_dict
```

```python
# Prepare data for modeling
y = df_enhanced['attack']
X = df_enhanced.drop(columns=['attack'], axis=1)

# Encode categorical variables using one-hot encoding
X_encoded = pd.get_dummies(X, drop_first=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

### Feature Selection

```python
# Train a Random Forest model to compute feature importance
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Select features with importance above a threshold
from sklearn.feature_selection import SelectFromModel
selector = SelectFromModel(rf_model, threshold="median", prefit=True)
X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)
```

```python
# Get the selected feature names
selected_features = X_encoded.columns[selector.get_support()]
selected_features
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Logistic Regression
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}
```

```python
# Initialize Logistic Regression model
log_reg = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)

# Perform Grid Search with Cross-Validation
grid_search = GridSearchCV(estimator=log_reg, param_grid=param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=1)
grid_search.fit(X_train_selected, y_train)
```

```python
# Output best parameters and best F1-score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best F1 Score:", best_score)
```

### Final Training with Optimized Parameters

```python
# Train the final Logistic Regression model using best parameters
final_log_reg = LogisticRegression(C=100, penalty='l1', solver='liblinear', class_weight='balanced', max_iter=1000, random_state=42)
final_log_reg.fit(X_train_selected, y_train)
```

```python
# Make predictions
y_pred = final_log_reg.predict(X_test_selected)
y_pred_proba = final_log_reg.predict_proba(X_test_selected)[:, 1]
```

```python
# Generate confusino matrix
confusion_matrix(y_test, y_pred)
```

```python
# Generate classification report
classification_rep = classification_report(y_test, y_pred)
print(classification_rep)
```

```python
# Calculating NEW predicted probabilities, false positive rates (FPR), true positive rates (TPR) and thresholds
y_probs_new = final_log_reg.predict_proba(X_test_selected)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_probs_new)
```

```python
# Calculating Area under Curve
roc_auc_new = auc(fpr, tpr)
roc_auc_new
```
